# 梯度下降法

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-1, 11, 141)
y = (x - 5)** 2
plt.plot(x, y)
plt.show()
```

```python
def dJ(theta):
    return 2*(theta-5)
```

```python
def J(theta):
    return (theta-5) ** 2 + 1
```

```python
eta = 0.1       #学习率
epsilon = 1e-8  #阈值

theta = 0.0
while True:
    gradient = dJ(theta)
    last_theta = theta
    theta = theta - eta * gradient
    
    if abs(J(theta) - J(last_theta)) < epsilon:
        break
print(theta)
print(J(theta))
```

绘制梯度下降过程

```python
eta = 0.000001
epsilon = 1e-6
theta = 0.0
theta_history = [theta]
while True:
    gradient = dJ(theta)
    last_theta = theta
    theta = theta - eta * gradient
    theta_history.append(theta)
    
    if abs(J(theta) - J(last_theta)) < epsilon:
        break

plt.plot(x, J(x))
plt.plot(np.array(theta_history), J(np.array(theta_history)), color = 'r', marker = 'o')
plt.show()
```

```python
len(theta_history)
```

- 学习率太小会收敛慢

```python
eta = 0.9
epsilon = 1e-6
theta = 0.0
theta_history = [theta]
while True:
    gradient = dJ(theta)
    last_theta = theta
    theta = theta - eta * gradient
    theta_history.append(theta)
    
    if abs(J(theta) - J(last_theta)) < epsilon:
        break

plt.plot(x, J(x))
plt.plot(np.array(theta_history), J(np.array(theta_history)), color = 'r', marker = 'o')
plt.show()
```

```python
print(theta_history[-1])
print(J(theta_history[-1]))
```

- 学习率过大，震荡，甚至不收敛

## 梯度下降法的优势

```python
import numpy as np
m = 1000   #样本数
n = 10000   #特征数

big_X = np.random.normal(size=(m,n))
true_theta = np.random.uniform(0.0, 100, size=n+1)
big_y = big_X.dot(true_theta[1:]) + true_theta[0] + np.random.normal(0., 10., size=m)
```

```python
from myML.LinearRegression import LinearRegression
reg1 = LinearRegression()
reg2 = LinearRegression()
```

```python
%time reg1.fit_normal(big_X, big_y)
```

```python
%time reg2.fit_gd(big_X, big_y)
```

- 可以看出，在特征数非常多的时候，使用梯度下降法比公式计算的实现要快的多

## 随机梯度下降法

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
m = 100000   #样本数
x = np.random.random(size=m)
y = 4. * x + 3 + np.random.normal(size=m)
print(x.shape)
print(y.shape)
```

```python
X = x.reshape(-1, 1)
X.shape
```

```python
from myML.LinearRegression import LinearRegression
reg1 = LinearRegression()    #样本数较多的时候，批量梯度下降法较慢
%time reg1.fit_gd(X, y)
```

```python
print(reg1.coef_)
print(reg1.interception_)
```

```python
reg2 = LinearRegression()    #样本数较多特征数较少的时候直接公式计算比较快
%time reg2.fit_normal(X, y)   
```

```python
print(reg2.coef_)
print(reg2.interception_)
```

#### 随机梯度下降法

```python
reg3 = LinearRegression()
```

```python
%time reg3.fit_sgd(X, y)
```

- 显然，在样本数较多特征数较少的时候随机梯度下降法比批梯度下降法要快，但是都比公式计算要慢很多，下面来看一下样本数和特征数都比较多的情况

```python
import numpy as np
m = 3000   #样本数
n = 3000   #特征数

big_X = np.random.normal(size=(m,n))
true_theta = np.random.uniform(0.0, 100, size=n+1)
big_y = big_X.dot(true_theta[1:]) + true_theta[0] + np.random.normal(0., 10., size=m)
```

```python
%time reg1.fit_normal(big_X, big_y)
```

```python
print(reg3.coef_)
print(reg3.interception_)
```

```python
%time reg2.fit_sgd(big_X, big_y, epoch=1)
```

```python
print(reg3.coef_)
print(reg3.interception_)
```

```python
%time reg3.fit_gd(big_X, big_y)  #时间非常长
```

```python
print(reg3.coef_)
print(reg3.interception_)
```

**总结：**
- 批梯度下降（损失下降比较稳定，样本数较少特征数较多的时候用）
- 随机梯度下降（损失下降不稳定，样本数较多的时候用）
- 公式实现（只有在特征数非常少的时候用）
