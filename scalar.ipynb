```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import datasets
```

```python
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
print(X.shape)
print(X[0])
print(y[0])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_trian, y_test = train_test_split(X, y)
```

### sklearn中的standardScaler

```python
from sklearn.preprocessing import StandardScaler
standardScalar = StandardScaler()
```

```python
standardScalar.fit(X_train)
```

```python
standardScalar.mean_
```

```python
standardScalar.scale_  #数据分布范围，标准差
```

```python
X_train_standarded = standardScalar.transform(X_train)
```

```python
X_train_standarded
```

```python
X_test_standard = standardScalar.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_clf = KNeighborsClassifier(n_neighbors=3, n_jobs=-1)
```

```python
knn_clf.fit(X_train_standarded, y_trian)
```

```python
knn_clf.score(X_test, y_test)
```

```python
knn_clf.score(X_test_standard, y_test)
```

可以看出归一化影响很大
